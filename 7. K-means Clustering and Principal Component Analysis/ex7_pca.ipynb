```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.io
%matplotlib inline
```

```python
data = scipy.io.loadmat('data/ex7data1.mat')
```

```python
X = data['X']
X[:5]
```

```python
plt.plot(X[:,0], X[:,1], 'bo')
```

```python
def featureNormalize(X):
    mu = X.mean(axis=0)
    sigma = X.std(axis=0)
    return (X - mu) / sigma ,mu, sigma
X_norm, mu, sigma = featureNormalize(X)
print 'mu: ' + str(mu)
print 'sigma: ' + str(sigma)
print X_norm[:5]
```

第一步，先標準化，上面做了  
第二步，計算協方差矩陣，就是共變異數矩陣$$\frac{1}{m}\sum_{i=1}^{n}(x^{(i)})(x^{(i)})^{T}$$shape會是n*n   
第三步，這個協方差矩陣的特徵向量

```python
def pca(X):
    sigma = 1. / X.shape[0] * np.dot(X.T, X)
    U, S, V = np.linalg.svd(sigma)
    return U, S
U,S = pca(X_norm)
print U
print S
```

```python
plt.figure(figsize=(5,5))
plt.plot(X[:,0], X[:,1], 'bo')
mu0 = mu + 1.5 * S[0] * U[:,0].T
mu1 = mu + 1.5 * S[1] * U[:,1].T
plt.plot([mu[0], mu0[0]], [mu[1], mu0[1]], '-k', lw=2)
plt.plot([mu[0], mu1[0]], [mu[1], mu1[1]], '-k', lw=2)
plt.show()
```

```python
plt.figure()
plt.scatter(X_norm[:, 0], X_norm[:, 1], marker='o', color='b', facecolors='none', lw=1.0)
plt.axis([-4, 3, -4, 3]) #axis square
plt.axis('equal')
plt.show()
```

只要用U的前K個向量，就可以得到方向向量
$$z^{(i)} = U^{T} \times x^{(i)}$$

```python
K = 1
def projectData(X, U, K):
    Z = np.zeros((X.shape[0], K))
    for i in range(X.shape[0]):
        u = U[:,:K].T
        Z[i] = u.dot(X[i])
    return Z
Z = projectData(X_norm, U, K)
Z[:5]
```

若要還原$$x^{i} = U\times z^{(i)}$$

```python
def recoverData(Z, U, K):
    X_rec = np.zeros((Z.shape[0], U.shape[0]))
    for i in range(Z.shape[0]):
        X_rec[i] = U[:,:K].dot(Z[i])
    return X_rec
X_rec = recoverData(Z, U, K)
print 'Recovery: ' + str(X_rec[:5])
print 'Original: ' + str(X_norm[:5])
```

```python
plt.figure(figsize=(5,5))
plt.scatter(X_rec[:, 0], X_rec[:, 1], marker='o', color='r', facecolor='none', lw=1.0, label='recovery')
plt.scatter(X_norm[:, 0], X_norm[:, 1], marker='o', color='b', facecolor='none', lw=1.0, label='original')
for i in range(len(X_norm)):
    plt.plot([X_norm[i, 0], X_rec[i, 0]], [X_norm[i, 1], X_rec[i, 1]], '--k')
plt.grid(True)
plt.legend(loc=0)
plt.show()
```
