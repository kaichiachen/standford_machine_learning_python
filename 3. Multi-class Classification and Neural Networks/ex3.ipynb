```python
Lambda = 0.1
```

跟之前分類的cost function 一樣

```python
def sigmoid(z):
    return 1 / (1 + math.e ** -z)
def lrCostFunction(theta, X, y, Lambda):
    H = sigmoid(X.dot(theta))
    #T = (-y* np.log(H)) - (1 - y) * (np.log(1 - H))
    J = np.sum(np.multiply(-y, np.log(H)) - np.multiply((1 - y), np.log(1-H))) / len(y) + Lambda / (2 * len(y)) * np.sum(theta[1:] ** 2)
    return J
def oneVsAll(X, y, num_labels, Lambda):
    X = np.column_stack((np.ones((X.shape[0],1)),X))
    all_theta = np.zeros((num_labels, X.shape[1]))
    for i in range(1,11):
        initial_theta = np.zeros(X.shape[1])
        options = {'full_output': True, 'maxiter': 50}
        theta, _, _, _, _ = optimize.fmin_cg(lambda t : lrCostFunction(t, X, 1*(y==i).flatten(), Lambda), initial_theta, **options)
        all_theta[i-1] = theta
        print i
    return all_theta
all_theta = oneVsAll(X, y, num_labels, Lambda)
```

```python
def predictOneVsAll(all_theta, X):
    X = np.column_stack((np.ones((X.shape[0],1)),X))
    C = sigmoid(X.dot(all_theta.T))
    print np.argmax(C,axis=1)+1
    return np.argmax(C,axis=1)+1
```
