### mapping object type feature and find which have many null column or some null column

```python
object_columns = data.select_dtypes(['object']).columns
non_null_columns_list = []
many_null_columns_list = []
some_null_columns_list = []
unique_less_than_50_sum = 0
some_null_sum = 0
many_null_sum = 0
for column in object_columns:
    if len(data[column].unique()) <= 50:
        unique_less_than_50_sum += 1
        if len(data[column][data[column].notnull()]) == len(price_val):
            unique_list = sorted(data[column].unique())
            mapping_val = dict(zip(unique_list, range(0, len(unique_list))))
            new_column = column + '_val'
            data[new_column] = data[column].map(mapping_val).astype(int)
            non_null_columns_list.append(column)
        elif len(data[column][data[column].notnull()]) >= len(price_val) * 0.8:
            some_null_sum += 1
            unique_list = sorted(data[column].unique())
            mapping_val = dict(zip(unique_list, range(0, len(unique_list))))
            new_column = column + '_val'
            data[new_column] = data[column].map(mapping_val).astype(int)
            some_null_columns_list.append(column)
        else:
            many_null_sum += 1
            unique_list = sorted(data[column].unique())
            mapping_val = dict(zip(unique_list, range(0, len(unique_list))))
            new_column = column + '_val'
            data[new_column] = data[column].map(mapping_val).astype(int)
            many_null_columns_list.append(column)
print 'object type column len: %d' % len(object_columns)
print 'unique_less_than_50_sum: %d' % unique_less_than_50_sum
print 'some_null_sum: %d' % some_null_sum
print 'many_null_sum: %d' % many_null_sum
```

```python
data_val = data.copy()
data_val = data_val.drop(non_null_columns_list,axis=1)
data_val = data_val.drop(many_null_columns_list,axis=1)
data_val = data_val.drop(some_null_columns_list,axis=1)
len(data_val.select_dtypes(['object']).columns)
```

```python
def clean_data(data):
    data_len = len(data)
    object_columns = data.select_dtypes(['object']).columns
    non_null_columns_list = []
    many_null_columns_list = []
    some_null_columns_list = []
    unique_less_than_50_sum = 0
    some_null_sum = 0
    many_null_sum = 0
    for column in object_columns:
        if len(data[column].unique()) <= 50:
            unique_less_than_50_sum += 1
            if len(data[column][data[column].notnull()]) == data_len:
                unique_list = sorted(data[column].unique())
                mapping_val = dict(zip(unique_list, range(0, len(unique_list))))
                new_column = column + '_val'
                data[new_column] = data[column].map(mapping_val).astype(int)
                non_null_columns_list.append(column)
            elif len(data[column][data[column].notnull()]) >= data_len * 0.8:
                some_null_sum += 1
                unique_list = sorted(data[column].unique())
                mapping_val = dict(zip(unique_list, range(0, len(unique_list))))
                new_column = column + '_val'
                data[new_column] = data[column].map(mapping_val).astype(int)
                some_null_columns_list.append(column)
            else:
                many_null_sum += 1
                unique_list = sorted(data[column].unique())
                mapping_val = dict(zip(unique_list, range(0, len(unique_list))))
                new_column = column + '_val'
                data[new_column] = data[column].map(mapping_val).astype(int)
                many_null_columns_list.append(column)
                
    data_val = data.copy()
    data_val = data_val.drop(non_null_columns_list,axis=1)
    data_val = data_val.drop(many_null_columns_list,axis=1)
    data_val = data_val.drop(some_null_columns_list,axis=1)
    
    for column in some_null_columns_list:
        column_val = column + '_val'
        mean = data_val[column_val].value_counts().argmax()
        data_val[column_val] = data_val[column_val].fillna(mean)
    data_val = data_val.drop([(column + '_val') for column in many_null_columns_list],axis=1)
    na_column = data_val.columns[pd.isnull(data_val).any()]
    for column in na_column:
        data_val[column] = data_val[column].fillna(data_val[column].mean())
    data_val = data_val.drop(['Id'],axis=1)
    
    return data_val
```

## Start training

```python
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier(n_estimators=100)
```

```python
data = pd.read_csv('train.csv')
train_X, train_y = clean_data(data)
```

```python
train_X.shape
```

```python
train_y.shape
```

```python
data = pd.read_csv('test.csv')
test_X, _ = clean_data(data)
```

```python
clf = clf.fit(train_X, train_y)
score = clf.score(train_X, train_y)
"Mean accuracy of Random Forest: {0}".format(score)
```

```python
test_y = clf.predict(test_X)
```
